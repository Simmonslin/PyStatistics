注意事項與討論 :

- 將數量的決定放在 code 第一條，譬如，n = 6

- 將符號的決定放在 code 第二條，譬如，marker = ‘s’

- 不論畫同心圓或圓上方塊符號，可以採用迴圈方式或非迴圈的矩陣計算方式。最好兩者都試試看（從自己最有把握的方法先做），才知道 python 的長處（可以先不論顏色）。

- 若要將標記標註為空心正方形，需將marker參數設為"s"，markfacecolor改為None

- 運用np.zeros()可將儲存空間限制於特定範圍，增加效率，否則電腦會自動擴增儲存空間，並降低程式碼跑速

- numpy.float64 並不支援index型式的分配 (e.g : [0][0] )，因此在寫np.zeros()時必須將其轉為list ( 可用.tolist()轉乘list)，否則會出現'numpy.float64' object does not support item assignment 的 error

- 用迴圈寫記號時需注意shape的正確性，否則可能會出現超過範圍的問題

```python
import numpy as np 
import matplotlib.pyplot as plt 

theta=np.linspace(0,2*np.pi,150)

a=2*5*np.sin(theta)
b=2*5*np.cos(theta)
c=2*10*np.sin(theta)
d=2*10*np.cos(theta)

print(a)


print(a[[10*i for i in range(10)]])




ax=plt.gca()

ax.grid(True)
ax.plot(a,b)
ax.plot([1,2,3,4],[0,0,0,0],marker="s")
ax.plot(c,d)
ax.set_aspect(1)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 


def sin_value():
    theta=np.linspace(0,2*np.pi,120+1)
    sin_v=np.zeros(6).tolist()
    for i in range(len(sin_v)):
        sin_v[i]=(10+2*i)*np.sin(theta)
        
    return sin_v

def cos_value():
    theta=np.linspace(0,2*np.pi,120+1)
    cos_v=np.zeros(6).tolist()
    for j in range(len(cos_v)):
        cos_v[j]=(10+2*j)*np.cos(theta)
    
    return cos_v 


def marker(x,y,num):
    marker_x=np.zeros(6).tolist()
    marker_y=np.zeros(6).tolist()
    
    for i in range(6):
        marker_x[i]=x[i][[10*i for i in range(num)]]
        marker_y[i]=y[i][[10*i for i in range(num)]]
        
    return marker_x, marker_y
    
    

ax=plt.gca()
ax.grid(True)

for line_num in range(len(sin_value())):
    ax.plot(sin_value()[line_num],cos_value()[line_num])
    
    
for line_num in range(len(sin_value())):
    ax.plot(marker(sin_value(),cos_value(),12)[0],marker(sin_value(),cos_value(),12)[1],marker="s",linewidth=2,linestyle="None",color="pink",markerfacecolor="None")
    
ax.set_aspect(1)

     
     
    
```
