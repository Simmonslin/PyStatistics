### 2. 繪製下圖（線條顏色與數量都可以由程式輕易決定）

<img src="https://ntpuccw.files.wordpress.com/2021/10/loop_proj.png" alt="getting started" width=400 />

注意事項與討論 :

- 需用到旋轉矩陣 $\begin{bmatrix}\cos\theta & \sin\theta \\-\sin\theta & \cos\theta \end{bmatrix}$ 讓圖案定點旋轉

- 設定 (0,0) (0,3) (-1,3) (-1,5) (1,3) (1,5) (1,3) (0,3) (0,0) 8個點形成圖案，並將其依 x , y 分成2個陣列

- 函式 rotate_matrix() 中的n=圖案數 ，若要產生n個圖案，則需要 8 x n 個 (x , y) 座標

- 為使無論多少個圖案皆能呈現360度的圖形，在設定旋轉矩陣時設定$\theta = \frac{2\pi}{n}$ 

- 每形成8個點，旋轉矩陣角度需要變換1次，因此用math.floor()函式 $\color{orange}(回傳小於等於數字的最大整數)$ 設定在8個點後，旋轉矩陣會進一位
  
  $\color{orange}(e.g : 從list_ROmatrix[0] -> list_ROmatrix[1])$

- x,y 陣列的index在迴圈中會遞增，此時用餘數(i%8)確保索引範圍在 0 - 7 之間

- 運用np.vstack()將 x,y 座標垂直堆疊後，再用倒置矩陣$\color{orange}(.T)$ 將矩陣變成 (2,n) 形式矩陣，最後取 x,y 值時，( [0] = x，[1] = y )

- np.vstack() 不接受 list 形態資料，因此需用 tuple() 將矩陣資料轉換成tuple型態

- 運用plt.subplot() 將不同圖合併 ( 211 代表 columns=2 , rows=1 , 最尾數 = 圖形擺放位置 )

```python
### 組裝方式  7個點拆開來 旋轉矩陣 * 單個點  --> 最後再用VSTACK+導致矩陣形成  2X7 矩陣 再進行plot



import numpy as np 
import matplotlib.pyplot as plt 
import math


x=np.array([0,0,-1,-1,1,1,0,0])
y=np.array([0,3,3,5,5,3,3,0])


#plt.plot(x,y)

def rotate_matrix(n):
    list_ROmatrix=np.zeros(n*8).tolist()
    list_product=np.zeros(n*8+1).tolist()
    
    for i in range(n*8):
        list_ROmatrix[i]=np.array([[np.cos(math.pi*(1/(0.5*n))*i),-np.sin(math.pi*(1/(0.5*n))*i)],[np.sin(math.pi*(1/(0.5*n))*i),np.cos(math.pi*(1/(0.5*n))*i)]])
        list_product[i+1]=np.dot(list_ROmatrix[math.floor(i/8)],np.array([x[i%8],y[i%8]]))
# rotate_matrix(n)[0-n]
      
    list_product=tuple(list_product)
    
    list_comb=np.vstack(list_product[1:]).T 
    
    return list_comb


plt.figure(figsize=(10,20))

plt.subplot(211).plot(rotate_matrix(8)[0],rotate_matrix(8)[1])
plt.subplot(212).plot(rotate_matrix(128)[0],rotate_matrix(128)[1],color="g")

    



```

注意事項與討論 :
 
 - 先設定 F 的範圍後，可用迴圈方式跑出不同自由度下的卡方值

 - 完成卡方分配表後，可用pd.DataFrame()將array轉換DataFrame，最後再用df.to_excel()將DataFrame轉換成excel檔
 
 - 可在pd.DataFrame()中加入$\color{orange}(columns)$ 和 $\color{blue}(index)$ 參數更改預設的index 和 columns

 - 可用sorted()將串列改成由大到小排列，index用迴圈重新設定

 - 轉換過程中發現.iloc[0,:]會自動變成NaN值

 - NaN值解決方法 : 將rows從29改成30行，原先.iloc[0,:]會變成空行，此時運用.dropna()將空值刪除

   結果 : 原先index為0的rows直接刪除，留下 rows 1-29 ，順利解決空行問題
 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2 
import pandas as pd 

area_upperTail=[0.005,0.01,0.025,0.05,0.1,0.9,0.95,0.975,0.99,0.995]



chi_table=np.zeros(30).tolist()

for i in range(len(chi_table)):
    chi_table[i]=chi2.ppf(area_upperTail,i)
    print(chi_table[i])
    
chi_table=pd.DataFrame(chi_table,columns=sorted(area_upperTail,reverse=True),index=[i for i in range (len(chi_table))]) 
chi_table=chi_table.dropna()
chi_table.to_excel("Chi2Table.xlsx")


print(chi_table)
    





 
```
