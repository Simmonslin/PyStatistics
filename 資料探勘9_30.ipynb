```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
data=np.array([200,300,400,600,1000])

scaler=StandardScaler()
scaled_data=scaler.fit_transform(data.reshape(-1,1))


df_ScaledData=pd.DataFrame(scaled_data,columns=['答案'])
print(df_ScaledData)
sns.distplot(df_ScaledData)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
data_MinMax=np.array([200,300,400,600,1000])

scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(data_MinMax.reshape(-1,1))


df_ScaledData=pd.DataFrame(scaled_data,columns=['答案'])
print(df_ScaledData)
sns.distplot(df_ScaledData)
plt.show()
```
