### 1. $ f(x)=\sin(x)+\cos(x) $

注意事項與討論 :
- 樣本數  $\color{green}( { np.linespace()中的num函數})$ 不能設太小，否則極值端點的弧度會不平滑

- 用 lambda 匿名函式可簡化程式碼

- 此數學式極值不超過 $\pm$ 1.5，因此 $y$ 區間不需設定太大範圍

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-2.5*np.pi,2.5*np.pi,400)
f= lambda x : np.sin(x)+np.cos(x)

ax=plt.gca()
ax.grid(True)
ax.set_xticks(np.array([-2.5,-2,-1.5,-1,-0.5,0,0.5,1,1.5,2,2.5])*np.pi)
ax.set_xticklabels(['-2.5$\pi$','-2$\pi$','-1.5$\pi$','-$\pi$','-0.5$\pi$','0','0.5$\pi$','$\pi$','1.5$\pi$','2$\pi$','2.5$\pi$']
                   ,fontsize=10,
                   color="green")
ax.plot(x,f(x))
ax.set_ylim([-2,2])
plt.title('$ f(x) = \sin(x)+\cos(x)$')
plt.show()
```

### 2. $ f(x) = \frac{1-\ell^{-2x}}{1+\ell^{-2x}} $

注意事項與討論 :
- 最大值為1，最小值為-1

- np.exp(1) 為自然常數本身，np.exp(2) 為自然常數的平方，因此 $\ell^{-2x}$ 應表達成 np.exp(-2*x) 而非 
  np.exp(-2**x)

```python
import numpy as np
import matplotlib.pyplot as plt
x2 = np.linspace(-10,10,2000)
f2 = lambda x2 : (1-np.exp(-2*x2))/(1+np.exp(-2*x2))

ax=plt.gca()
ax.set_ylim([-1.2,1.2])
ax.plot(x2,f2(x2))
ax.grid(True)
plt.scatter(0,0,s=70,color="red")
plt.title(r"$ f(x) = \frac{1-\ell^{-2x}}{1+\ell^{-2x}} $")
plt.annotate("(0,0)",(0.5,0),fontsize=15)

plt.show()
```

### 7. $ f(x) = 2x^{3}-x^{4}$

### 10. $x^{2}+y^{2}=1$

注意事項與討論 :

- 使用 $\sin\theta^{2}+\cos\theta^{2}=1$ 解題

- 承上，因此$x$ 和 $y$ 的區間設定在 0- $2\pi$ 之間 $\color{green}( 使用np.linspace() )$

- 使用 ax.set_aspect ( ) 將座標軸設定成正方形，圖形才能變得較圓

- 使用 math 模組取得 $ \pi $ 值

```python
import numpy as np 
import matplotlib.pyplot as plt
import math


x10=np.sin(np.linspace(0,math.pi*2,100))
y10=np.cos(np.linspace(0,math.pi*2,100))


ax=plt.gca()
ax.grid(True)
ax.set_aspect(1) 
ax.plot(x10,y10,linewidth=1)
plt.vlines(0,-1.5,1.5,color="red",linestyles="--")
plt.hlines(0,-1.5,1.5,color="red",linestyle='--')
plt.scatter(0,0,s=70,color="green")
plt.annotate("(0,0)",(0,0),fontsize=15,color="green")
plt.title('$ x^{2}+y^{2}=1$')
plt.show()
```

### 11. 繪製正方形

注意事項與討論 :


- 使用matplotlib 中的 patches function 繪製正方形 $\color{red}(matplotlib.patches)$

- add_patch() 使用方式

  1. 以左下角為起始點設定長度和寬度

  2. 取消實心方法 :將 facecolor參數設為 None
  
  
- 使用add.set_aspect() 將座標軸設定成正方形，使圖形正方化


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as pt 
import math


ax=plt.gca()
ax.set_xlim([-1.5,1.5])
ax.set_ylim([-1.5,1.5])
ax.add_patch(pt.Rectangle((-0.5,-0.5),1,1,edgecolor="blue",linewidth=2,facecolor="none",angle=math.degrees(math.pi/2)))
ax.vlines(0,-1.5,1.5,color="red",linestyle="--")
ax.hlines(0,-1.5,1.5,color="red",linestyle="--")
ax.scatter(0,0,s=70,color="green")
ax.annotate("(0,0)",(0,0),fontsize=15)
ax.set_aspect(1)
plt.show()

```
