- ECDF ( empirical distribution function ) = 經驗分布函數

  1. ECDF 圖來自樣本數為 n 的樣本，當 n 愈大時，與真實 CDF 愈接近。因此 ECDF 為真實 CDF 的模擬版本。<br><br>
  2. 模擬 CDF 最簡單的方式就是將每個樣本點的發生機率設為一樣

```python
from pyexpat.errors import XML_ERROR_TEXT_DECL
from matplotlib.lines import lineStyles
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import norm 

n=10000
x = norm.rvs(loc=0,scale=1,size=n)

# 如何確定 x 為常態分配  1. 畫 hist 
plt.hist(x,density=True,bins=50)

xx=np.linspace(-5,5,n)
norm_pdf=norm.pdf(xx)
plt.plot(xx,norm_pdf,lw=3)
plt.show()

x_sort=np.sort(x) 
ecdf=np.arange(1,n+1) / n
plt.plot(x_sort,ecdf)

norm_cdf=norm.cdf(xx)
plt.plot(xx,norm_cdf,alpha=0.8,linestyle="-.",lw=3,color="tomato")
plt.title("")
plt.show()



```

```python
from pyexpat.errors import XML_ERROR_TEXT_DECL
from matplotlib.lines import lineStyles
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import norm ,chi2

n=1000
x = norm.rvs(loc=0,scale=1,size=n)
y= x**2
# 如何確定 x 為常態分配  1. 畫 hist 
plt.hist(y,density=True,bins=100)
plt.title("$x^{2} = 1$")

xx=np.linspace(0,5,n)
chi2_pdf=chi2.pdf(xx,df=1)
plt.plot(xx,chi2_pdf)
plt.show()

x_sort=np.sort(x) 
ecdf=np.arange(1,n+1) / n
plt.plot(x_sort,ecdf)

norm_cdf=norm.cdf(xx)
plt.plot(xx,norm_cdf,alpha=0.8,linestyle="-.",lw=3,color="tomato")
plt.show()
```

#### 如何在 x ( -1,1 )  y = ( -1,1 ) 的正方形中 ，均勻撒下 1000 個點 ? 


```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt

n = 1000
N = 1000

pi_hat=np.zeros(1000)
for i in range(N):
    x = uniform.rvs(loc=-1, scale=2, size=n)
    y = uniform.rvs(loc=-1, scale=2, size=n)
    R = np.sqrt(x**2+y**2)
    # M = 算 True / False 數量
    M = (R < 1).sum()
    pi_hat[i] = 4*M/n

plt.hist(pi_hat)

print(pi_hat.mean())
```
